```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 讀取資料集
data = pd.read_csv('Data/recorded_data-4.txt', sep='\t')

data['time'] = pd.to_datetime(data['time'])
data['time'] = data['time'].dt.strftime('%H:%M:%S')
data_subset = data.iloc[100:]
data_subset.info()
data.head()
```

```python
# 創建 2D 密度圖
sns.kdeplot(x=data_subset['position_x'], y=data_subset['position_y'], cmap="Oranges", fill=True, thresh=0)

# 加入軸標籤和標題
plt.xlabel('Position X')
plt.ylabel('Position Y')
plt.title('Heatmap of Screen Positions')

plt.gca().invert_yaxis()
# 顯示熱力圖
plt.show()
```

```python
sns.set_theme(style="whitegrid")

# 創建邊線柱狀圖
sns.histplot(data_subset, x='position_x', y='position_y', cbar=True, cmap="Blues", cbar_kws={'label': 'Density'})

# 加入軸標籤和標題
plt.xlabel('Position X')
plt.ylabel('Position Y')
plt.title('Heatmap of Screen Positions')
plt.gca().invert_yaxis()
# 顯示圖形
plt.show()
```
